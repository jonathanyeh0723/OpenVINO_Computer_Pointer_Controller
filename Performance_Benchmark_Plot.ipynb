# Define Plot Function

```python
# Import plot libraries
import matplotlib.pyplot as plt
import numpy as np

# Plot Function
def plot_func(value, measurement, color_type = "b"):
    plt.xlabel("Model Precision")
    plt.ylabel("Seconds")
    plt.title(measurement)
    plt.bar(["FP32", "FP16", "FP16-INT8"], value, color = color_type)
```

## Model Load Time

```python
value = [0.386, 0.429, 0.800]
measurement = "Model Load Time"

# Plot
plot_func(value, measurement)
```

## Inference Time

```python
value = [24.175, 25.024, 24.467]
measurement = "Model Load Time"

# Plot
plot_func(value, measurement, color_type = "g")
```

## Frame Per Second

```python
value = [2.441, 2.358, 2.411]
measurement = "Frame Per Second (FPS)"

# Plot
plot_func(value, measurement, color_type = "brown")
```
